```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Assignment_Excel.xlsx", sheet_name=0)
# insert the name of the column as a string in brackets
list1 = list(df['Input String']) 
list2 = list(df['Output String'])

print(list1)
print(list2)
```

```python
output_list = []
```

```python
for s in list1:
  new_s = s.split(' ')
  print(new_s)
  #out_s=[]
  i=0
  while new_s[i]=='\n' or new_s[i]=='\n\n':
    i=i+1
  #print(new_s[i])
  new_s[i] = new_s[i].title()
  #print(new_s[i])
  res = ' '.join(new_s)
  print(res)
  #print(out_s)
  output_list.append(res)
```

```python
# Generate dataframe from list and write to xlsx.
pd.DataFrame(output_list).to_excel('output.xlsx', header=False, index=False)
```
